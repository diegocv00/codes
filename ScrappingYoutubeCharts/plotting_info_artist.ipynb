```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
archivo = "top10_artistas_detalle.xlsx"
xls = pd.ExcelFile(archivo)

# Mostrar todas las hojas disponibles
print(xls.sheet_names)
```

```python
# Leer la hoja de visitas de "Bad Bunny"
df_visitas = pd.read_excel(archivo, sheet_name="Bad Bunny_visitas")

# Leer la hoja de ciudades de "Bad Bunny"
df_ciudades = pd.read_excel(archivo, sheet_name="Bad Bunny_ciudades")

# Leer la hoja de canciones de "Bad Bunny"
df_canciones = pd.read_excel(archivo, sheet_name="Bad Bunny_canciones")
```

GRAFICANDO LAS VISITAS DIARIAS

```python
plt.figure(figsize=(20,10))
df_visitas["Dia_Mes"] = df_visitas["Fecha"].str[:6] 
mes = df_visitas["Fecha"].str[7:11]
sns.lineplot(data=df_visitas,x="Dia_Mes",y="Visitas")
plt.xlabel(f"Fecha({mes[0]})")
```

GRAFICANDO LAS CIUDADES QUE MAS ESCUCHAN AL ARTISTA

```python
plt.figure(figsize=(16,10))
df_ciudades["Visitas"] = df_ciudades["Visitas"].str[:4]
df_ciudades["Visitas"] = df_ciudades["Visitas"].astype(float)
df_ciudades.rename(columns={"Visitas":"Visitas(en millones)"},inplace=True)
sns.barplot(data=df_ciudades,x="Ciudad",y="Visitas(en millones)")
plt.xticks(rotation=45)
```

GRAFICANDO LAS CANCIONES MAS ESCUCHADAS DEL ARTISTA

```python
plt.figure(figsize=(16,10))
df_canciones["Visitas"] = df_canciones["Visitas"].str[:4]
df_canciones["Visitas"] = df_canciones["Visitas"].astype(float)
df_canciones.rename(columns={"Visitas":"Visitas(en millones)"},inplace=True)
sns.barplot(data=df_canciones,y="Canción",x="Visitas(en millones)")
```
